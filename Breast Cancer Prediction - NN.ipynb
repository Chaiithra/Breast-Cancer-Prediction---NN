```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Separating the Features and the Target

```python
x = df.drop(columns = 'label', axis = 1)
y = df['label']
```

```python
x
```

```python
y
```

Splitting the data into training & testing data

```python
xtrain, xtest, ytrain, ytest = train_test_split( x, y, test_size = 0.2, random_state = 2)
```

```python
print(x.shape, xtrain.shape, xtest.shape)
```

Standardize the data 

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

xtrain_std = scaler.fit_transform( xtrain )

xtest_std  = scaler.transform( xtest )
```

```python
print( xtrain_std )
```

Building the Neural Network 

```python
# Importing the required libraries

import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras 
```

```python
# Setting up the layers of Neural Network

model = keras.Sequential([
    
    keras.layers.Flatten(input_shape=(30,)), # Here data should be inputted as 1-D Array
    keras.layers.Dense(20, activation = 'relu'),
    keras.layers.Dense(2, activation = 'sigmoid')
    
])
```

```python
# Compiling the Neural Network

model.compile( optimizer = 'adam',
             loss = 'sparse_categorical_crossentropy',
             metrics = ['accuracy'])
```

```python
# Training the Neural Network

history = model.fit(xtrain_std, ytrain, validation_split = 0.1, epochs = 10)
```

Visualizing accuracy and loss 

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')

plt.xlabel('epochs')
plt.ylabel('accuracy')

plt.legend(['training data','validation data'], loc = 'lower right' )
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')

plt.xlabel('epochs')
plt.ylabel('loss')

plt.legend(['training data','validation data'], loc = 'upper right' )
```

Accuracy of the model on the test data

```python
loss, accuracy = model.evaluate( xtest_std, ytest)
```
